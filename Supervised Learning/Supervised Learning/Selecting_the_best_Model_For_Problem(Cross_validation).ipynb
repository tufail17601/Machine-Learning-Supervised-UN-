```python
x = df.drop(columns=['prognosis', 'Unnamed: 133'],axis = 1)
```

```python
y = df['prognosis']
```

```python
df['prognosis'].nunique
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
y
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=200),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Machine': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
}
```

```python
result = []
for name, model in models.items():
    acc = cross_val_score(model, x, y, cv=cv, scoring='accuracy').mean()
    prec = cross_val_score(model, x, y, cv=cv, scoring='precision_macro').mean()
    rec = cross_val_score(model, x, y, cv=cv, scoring='recall_macro').mean()
    f1 = cross_val_score(model, x, y, cv=cv, scoring='f1_macro').mean()
    result.append({
        'Model': name,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1': f1
    })
p_result = pd.DataFrame(result)
    
```

```python
p_result
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```
